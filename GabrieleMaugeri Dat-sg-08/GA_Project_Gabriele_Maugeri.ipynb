# Import Libraries for Naive Bayes

```python
import scipy as sp
import random

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression 
from sklearn import metrics

from textblob import TextBlob, Word

from nltk.stem.snowball import SnowballStemmer

%matplotlib inline
```

```python

#Creation of a DataFrame that contains only 5-stars and 1-stars reviews
Top_Low_Rev = Rev[(Rev.Rating==5) | (Rev.Rating==1)]

Top_Low_Rev.head()
```

```python

#Shuffle of the DataFrame
Top_Low_Rev = Top_Low_Rev.sample(frac=1)


#Define X and y
X = Top_Low_Rev.Content
y = Top_Low_Rev.Rating


Top_Low_Rev.head()
```

```python

#Splitting of the DataFrame in 2 parts: Training DataFrame and Testing DataFrame
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
```

```python
#Rating Model using CountVectorizer and Naive Bayes


#Use of CountVectorizer
vect = CountVectorizer()

# Create document-term matrices.
X_train_dt = vect.fit_transform(X_train)
X_test_dt = vect.transform(X_test)

vect
```

```python

# Apply Naive Bayes model to predict the Stars Rating
nb = MultinomialNB()
nb.fit(X_train_dt, y_train)
y_pred = nb.predict(X_test_dt)

# Calculate accuracy.
print((metrics.accuracy_score(y_test, y_pred)))
```

```python
#Control Matrix
y_test.value_counts()
```

```python
# Calculate null accuracy.
y_test_binary = np.where(y_test==5, 1, 0) # five stars become 1, one stars become 0
print('Percent 5 Stars:', y_test_binary.mean())
print('Percent 1 Stars:', 1 - y_test_binary.mean())
```

```python
# Real accuracy of the model over the null
print("For 5 Stars reviews the model is ", (((metrics.accuracy_score(y_test, y_pred)) - (y_test_binary.mean()))*100),"% more accurate than null accuracy")
print("For 1 Stars reviews the model is ", (((metrics.accuracy_score(y_test, y_pred)) - (1 - y_test_binary.mean()))*100),"% more accurate than null accuracy")
```

We can say that the Model predicted ~94% accuracy, which is an improvement over this baseline 86% accuracy (assuming our model always predicts 5 stars).
